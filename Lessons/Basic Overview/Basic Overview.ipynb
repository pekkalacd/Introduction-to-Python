Today, the world runs on computers. Everything, from businesses needing to handle their internal data and customer records, to finance personnell needing to complete transactions on the stock market, to massive tech companies innovating and building consumer products that are shaping the way we see life, relies heavily on the presence of computers.

**What is a computer?**

A computer is a machine that can be instructed to carry out sequences of arithmetic and logical operations via computer programs.

<br>

**What is a computer program?**

As hinted at before, a computer program is a sequence of arithmetic and logical statements that connect together into specific instructions to get a computer to perform a specific task. In order to write a program, ***programmers*** (like you!) use a ***programming language*** in order to communicate these instructions to the computer. But computers do not understand the programming language alone. Instead, they acknowledge what is known as ***machine language*** or 1's and 0's to control physical processes such as firing on and off set voltages through a series of transistors. 

<br>

**From programming language to machine language**

If programmers use programming languages to communicate with computers, that are not 1's and 0's, and computer's only understand 1's and 0's, what happens in between the programmer and the computer? Striking question! The languages that programmers use are higher-level abstractions from what is literally happening within the machine. Those 1's and 0's (or **binary**) that are being produced are put into specific streams of sequences representing the transfer of data from one location in memory to another within the system itself, along with specific actions such as utilizing built-in storage spaces and ports known as peripherals on the hardware itself. This process of sending streams of binary from a set of ***source code*** (programming language) into machine language is complex and requires several intermediate languages known collectively as ***assembly*** to break down the logic presented by the programmer into lower-level chunks that can be easily translated into machine language. 

An assortment of programming languages use something called a *compiler*. A compiler is a piece of software which translates high-level source code into machine language. For example, in languages like C and C++, the process from taking a text file into an executable utilized the ***preprocessor***, ***compiler***, and ***linker***. The preprocessor is a piece of software which processes a set of commands known as ***directives***. These directives perform purely textual manipulation such as including sets of source code not found on the native system into a source code file. The preprocessor then removes the directives and returns a ***modified source code*** file, which contains strictly source code, into the compiler. The compiler condenses the source code into machine language by utilizing the assembler, which moves the source code into assembly and then from assembly into machine language. When the compiler is finished with this process an ***object code*** file is created. The object code file contains almost executable binary, but not quite. The machine language does not contain the previously accounted for non-native source code found by the preprocessor. The linker links the header files (***libraries***) accounted for by the preprocessor to the machine language contained within the object code file, in the appropriate places, and returns an executable file. 

Today, many programmers use something known as an ***Integrated Development Environment*** (IDE), to do their work. For compiled languages, most IDEs, will complete that entire afore mentioned process with the click of a button!

**What is Python?**

Python is a general purpose object-oriented programming language. Though, traditionally it has been thought of as a ***scripting language***, Python has grown into in a stand alone language in many applications. Unlike C and C++, Python is <u>not a compiled language</u>.  It is an ***interpreted language***. The main benefit to this is evaluating programs at ***run-time*** vs ***compile-time***. A compiled language requires that a program must be written to abide by the syntax as guided by the language, to avoid syntax and compile-time errors, in order to run. Whereas, an interpreted language, can be immediately run line by line. This makes interpreted languages, generally, easier to test, find bugs, and resolve errors. Match that with a simple syntax and a plethora of built-in functions, that can handle plenty of tasks, and you have Python. An easily testable, writable, understandable, interpreted language that can do a lot. 

<br>

**Brief History of Python**

In the mid-1980s, Guido Van Rossum was a developer for the ABC system (an imperative general purpose programming language aimed at teaching and prototyping). Later, in 1989, Van Rossum started a hobby side project, a new scripting language aimed at appealing to the ABC and UNIX/C hacker community. A fan of Monty Python's Flying Circus, Guido attributed the name "Python" to this language. 

In 1999, Van Rossum submitted a "Computer Programming for Everyone" proposal to DARPA where he defined the following goals: 
- An easy open source intuitive language just as powerful as major competitors
- Open source, so that any one can contribute to it's development
- Code that is as understandable as plain English
- Suitability for everyday tasks, allowing for short development times

Since 2004, Python has been among the 10 most popular programming languages every year. In the past 3 years, Python was among the top 3 most-used languages on GitHub.

**Why learn Python?**

Python is everywhere! Google, Twitter, DropBox, Facebook, etc. all at some level use Python. Typically, it is used on the ***back end*** of their varying applications, to handle things such as ***requests*** to a server, ***querying*** and ***inputting*** data from and to databases, and generating ***authentication tokens*** for security. 

Arguably, it's most notable application, is in the ***data analysis and visualization*** space. Where a group of known ***libraries*** such as [**pandas**](https://pandas.pydata.org/), [**seaborn**](https://seaborn.pydata.org/), [**numpy**](https://numpy.org/), [**matplotlib**](https://matplotlib.org/), [**tensorflow**](https://www.tensorflow.org/guide/data), [**pytorch**](https://pytorch.org/), and [**more**](https://www.geeksforgeeks.org/top-10-python-libraries-for-data-science-in-2020/) reside. 

Python also has a space in web development. Frameworks such as [**Flask**](https://flask.palletsprojects.com/en/1.1.x/) and [**Django**](https://www.djangoproject.com/) can be used to build light-weight and traffic-bearing websites & applications. Companies like Instagram, Mozilla, Pinterest, National Geographic, and more use Django. Whereas, Companies like Red Hat, Rackspace, Airbnb, Netflix, Lyft, and Reddit use Flask.

Boasting more than 250,000 packages, tackling tasks like building games, websites, web apps, data science, data visualization, security, and more, Python has a ton to [**offer**](https://pypi.org/).

**How do you get Python?**

Python can be downloaded from the python website available at: [**https://www.python.org/downloads/**](https://www.python.org/downloads/). Make sure to download the latest version of Python, at the time of writing this it's 3.9.1. This tutorial will work for versions 3.8.0 or later.

Another tool used is known as a ***package manager***. We'll go into this in a much later lesson in more detail. But for now, if you'd like to install one, this tutorial will be using *pip* which should already be installed in your current version of Python if later than 3.4. Another notable package manager is [**conda**](https://docs.conda.io/projects/conda/en/latest/user-guide/install/windows.html).
